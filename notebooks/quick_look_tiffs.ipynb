```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
stem = '/Users/eckhartspalding/Documents/git.repos/nice/data/20260322_and_24_eckhart/'
```

```python
def get_frame(file_name):
    '''
    Return frame as a np.array
    '''
    bright_file_name = file_name
    bright_frame = Image.open(bright_file_name)
    bright_frame = np.array(bright_frame, dtype=np.float32)

    return bright_frame
```

```python
test_frame_hypercal_blank = get_frame(file_name = stem + 'hypercal_subted_frame.tiff')
test_frame_standard_bckgrnd_blank = get_frame(file_name = stem + 'standard_bckgrnd_subted_frame.tiff')
test_frame_hypercal_signal = get_frame(file_name = stem + 'hypercal_subted_150_mW.tiff')
test_frame_standard_bckgrnd_signal = get_frame(file_name = stem + 'standard_subted_150_mW.tiff')
```

```python
print(np.std(test_frame_hypercal))
print(np.std(test_frame_standard_bckgrnd))
```

```python
plt.imshow(test_frame_hypercal_signal)
plt.colorbar()
plt.title('hypercal subtracted')
plt.show()

plt.imshow(test_frame_standard_bckgrnd_signal)
plt.colorbar()
plt.title('standard bckgrnd-subtracted')
plt.show()
```
